```python
def levelOrder(root): 
    level_list =[]
    
    if not root:
        return level_list
    
    queue =deque()
    queue.append(root)
    
    level =0
    while queue:
        i =0
        n =len(queue)
        
#       ----------------每一层的处理----------------------
#         level_list.append([])
        tmp_list =[]
        while i<n:
            node =queue.popleft()
#             level_list[level].append(node.val)
            tmp_list.append(node.val)
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
            i+=1         
        level+=1
#       --------------------------------------
        level_list.append(tmp_list)
    return level_list,level

```

### BST LCA(last common ancestor)

```python
def LCA(root,p,q):   
    # Value of current node or parent node.
    parent_val = root.val

    # Value of p
    p_val = p.val

    # Value of q
    q_val = q.val
    
# 思路
# 1.两个点都在root左边，SET root=root.left
# 2.两个点都在root右边，SET root=root.right
# 3.两个点都在root两边，return root

    # If both p and q are greater than parent
    if p_val > parent_val and q_val > parent_val:    
        return self.lowestCommonAncestor(root.right, p, q)
    # If both p and q are lesser than parent
    elif p_val < parent_val and q_val < parent_val:    
        return self.lowestCommonAncestor(root.left, p, q)
    # We have found the split point, i.e. the LCA node.
    else:
        return root
```

### BT最小深度
⚠️ 和最大深度不一样，有可能是单边树，当其中一边为空的时候，就要考虑整个的深度

```python
def mindepth(root):
# 1.END  
    if not root:
        return 0

# 2.Recursive     
    l =mindepth(root.left)
    r =mindepth(root.right)
    
    if not root.left and root.right:
        return r+1
    if not root.right and root.left:
        return l+1
    
    
    depth =min(l,r)+1
    return depth
```

### PATH_record（根到叶节点）

```python
def all_path(root):
    res =[]
    
    def backtrack(path,root):
# 1.END  
        if not root:
            return 
       
    
# 2.HIT the Target         
# 2.穷尽选择（走到叶节点），就把这条 路径 放入结果中
        if not root.left and not root.right:
            res.append(path+'>'+str(root.val))
            return

# 3.Recursive  
# 3.还有选择，就recurse左右两边
        else:           
            backtrack(path+'>'+str(root.val),root.left)
            backtrack(path+'>'+str(root.val),root.right)

    backtrack('',root)
    return res
```

### Target(BOOLEAN) 路径和（根到叶节点）

```python
def haspathSum(root,target):
# 1.END  
    if not root:
        return False
    
# 2.HIT the Target 
    if root.val==target:#   once find one path, return True
        return True

# 2.Recursive 
    return haspathSum(root.left,target-root.val) or haspathSum(root.right,target-root.val)
```

### 二叉树的直径 DFS
[一个节点到另一个节点 中间有几个edge，最长的那个]

```python
class T:        
    def diameter(self,root):
        self.diam=0
        self.DFS(root)
        return self.diam
    
    # ==============💁超级重要==============
    def DFS(self,root):
# 1.END  
        if not root:
            return 0        
# 2.Recursive         
# 求树的深度=max(左，右)
#       这里是自底向上，所以每一个结点 都会被考虑到
        l =self.DFS(root.left)
        r =self.DFS(root.right)

        
# 求一个节点有几个边=左+右  
        self.diam=max(self.diam,l+r)
        
        return max(l,r)+1
    # ====================================
   
```

#### 1.非递归

```python
def f1(nums,target):
    if len(nums)==0:
            return -1
    if len(nums)==1:
        if nums[0]==target:
            return 0
        else:
            return -1
    start =0
    end =len(nums)-1
    while start<=end:
    
        mid =start+(end-start)//2
        if nums[mid]==target:
            return mid  
       
        if start>=end:      
            return -1 

        if nums[start]<nums[mid]:
            if nums[mid-1]>=target>=nums[start]:  
                end =mid-1
            else:
                start =mid+1
           

        else:
            if nums[mid+1]<=target<=nums[end]:
                start =mid+1          
            else:
                end =mid-1
        
    return -1
```

```python
nums=[2,0]
target=0
f1(nums,target)
```

###  开根

```python
def DIY_pow(a,b):
    l =0
    h =a
    res=(l+h)/2
    while abs(res**b-a)>0.01 and a>=1:
        if a==1:
            return 1.0000000
        elif a>1:
            if res**b<a:
                l =res
            else:
                h =res
            res =(l+h)/2
            
    while abs(res**b-a)<0.01 and a<=1:
            if res**b>a:
                h =res
            else:
                l =res
            res =(l+h)*2
    return res 
```

```python
def myPow(x,n):
    if n < 0:
        x = 1/x
        n = -n

    if n == 0:
        return 1
    if n == 1:
        return x

    if n % 2 == 0:
        return myPow(x, n/2)**2
    else:
        return myPow(x, n-1) * x
```
